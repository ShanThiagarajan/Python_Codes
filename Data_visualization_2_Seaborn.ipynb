## Data Visualization using Seaborn Library


```python
## Importing the library
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### Strip Plot

```python
## Loading the required data
df_titanic = pd.read_csv('datasets/Visualization/titanic_data.csv')
df_titanic.head()
```

```python
df_titanic.info()
```

```python
df_titanic.set_index('PassengerId')
```

```python
df_titanic.describe()
```

```python
## Strip plot for age and gender
sns.stripplot(x = 'Sex', y = 'Age', data = df_titanic)

plt.title('Strip plot for age and Gender')
plt.show()
```

```python
## Adding 'Survived' category variable to differentiate the surviors
## Strip plot for age and gender
sns.stripplot(x = 'Sex', y = 'Age', hue = 'Survived', data = df_titanic)

plt.title('Strip plot for age and Gender')
plt.show()
```

```python
df_titanic[df_titanic['Embarked'].isnull()]
```

```python
df_titanic['Embarked'].isnull().sum()
```

```python
df_titanic['Embarked'].fillna('C',inplace= True)
```

```python
sns.stripplot(x='Embarked',y = 'Age',hue='Survived',data = df_titanic)
plt.title("The Survivor Analysis by Port of Embarcation")
plt.show()
```

### Distribution Plots

```python
## Loading the required data
df_ins = pd.read_csv('datasets/Visualization/insurance_data_with_day.csv')
df_ins.head()
```

```python
df_ins['log_claim'] = np.log(df_ins['claim'])
df_ins.head()
```

```python
## Creating distribution plot
sns.displot(df_ins['claim'], color='teal',kde=True,bins=10)

plt.title('Distribution of Insurance Claim')
plt.show()
```

```python
## Creating distribution plot
sns.displot(df_ins['log_claim'], kde=True,bins=10)

plt.title('Distribution of Insurance Claim')
plt.show()
```

```python
## Creating distribution plot
sns.displot(df_ins['claim'], color='teal', kde=False, bins=10)

plt.title('Distribution of Insurance Claim')
plt.show()
```

### Joint Plot

```python
print("Smoker's min {} and max {} claims".format(df_ins[df_ins.smoker=='Yes']['claim'].min(),\
                                                 df_ins[df_ins.smoker=='Yes']['claim'].max()))
print("Non-smoker's min {} and max {} claims".format(df_ins[df_ins.smoker=='No']['claim'].min(),\
                                                 df_ins[df_ins.smoker=='No']['claim'].max()))
```

```python
df_ins.groupby('smoker')['claim'].agg([min, max, np.mean]).round(2)
```

```python
## Creating joint plot for BMI and Insurance Claim
sns.jointplot(x='bmi', y='claim', hue='smoker', data = df_ins, color='sandybrown')
plt.xlabel('BMI')
plt.ylabel("Claim")
plt.show()
```

```python
## Value Counts in Numbers
df_ins.smoker.value_counts()
```

```python
## Value Counts in Percentage (Normalize = True)
df_ins.smoker.value_counts(normalize=True).mul(100).round(2)
```

```python
df_ins.smoker.value_counts(normalize=True).mul(100).round(2).astype(str)+'%'
```

### Violin Plot

```python
df_titanic.describe()
```

```python
## Creating violin plot age and sex of titanic data set
sns.violinplot(x='Sex', y='Age', data=df_titanic)
plt.title("Distribution of Age and Gender using Violin Plot")
plt.show()
```

### Swarm Plot

```python
## Creating violin plot age and sex of titanic data set
sns.swarmplot(x='Sex', y='Age', data=df_titanic,size=4,  hue='Survived')
plt.title("Distribution of Age and Gender using Swarm Plot")
plt.show()
```

### Count Plot

```python
round((df_titanic.Sex.value_counts()[0]/df_titanic.shape[0]) * 100,2)
```

```python
sex_vc_percent = df_titanic.Sex.value_counts(normalize=True).mul(100).round(2)
print(sex_vc_percent)
print(type(sex_vc_percent))
```

```python
sex_vc_counts=df_titanic.Sex.value_counts()
sex_vc_counts
```

```python
## Creating Count Plot for sex
sns.countplot(x='Sex', data=df_titanic)

plt.text(x=-.1, y=sex_vc_counts['male']+10, s = str(sex_vc_percent['male'])+ '%'  )
plt.text(x=.9, y=sex_vc_counts['female']+10, s = str(sex_vc_percent['female'])+ '%'  )
  
plt.show()
```

### Heatmap

```python
## Will be displaying the iris data set for the correlation among the variables
import numpy as np
corr = df_iris.corr()

mask = np.triu(np.ones_like(corr, dtype=bool))

sns.heatmap(data = df_iris.corr(), mask=mask, center=0, annot = True, linewidth=0.5, cmap='Greens', cbar=True, vmin=-1, vmax=1)

## Correlation heatmap for iris dataset
plt.title("Correlation heatmap for iris dataset")

plt.show()
```

## LMPlot

```python
## Linear Regression Plot
## It displays the regression line for each category of data.
sns.lmplot(x='bmi',y='claim',hue='smoker',data=df_ins)
plt.show()
```

## Category Plot

```python
## Default Strip Plot
sns.catplot(x='region',y='claim',hue='smoker',data=df_ins)
plt.show()
```

```python
## Violin Categorical Plot
plt.figure(figsize=(20,15))
sns.catplot(x='region',y='claim',kind='violin',data=df_ins)
plt.show()
```
